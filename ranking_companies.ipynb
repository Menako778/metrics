```python
from openai import AsyncOpenAI
import asyncio

model = ''
url = ''


async_model_client = AsyncOpenAI(base_url=url, api_key=ZELIBOBA_TOKEN)
```

```python
df_clustered = pd.read_csv('topic_ids')
```

```python
prompt_template = '''
You are an expert gaming analyst.  
Carefully analyze each game in the list below and generate a precise cluster name that best categorizes all games collectively.  

**Requirements:**  
1. The cluster name must be specific to the shared characteristics of ALL games in the list  
2. Avoid generic terms like "Action" or "Adventure" unless they are truly defining  
3. Consider genre hybrids, art style, gameplay mechanics, and target audience  

**Output Format:**  
Provide only the cluster name in bold without additional explanations.  

**List of Games:**  
{list_of_games}  
'''
```

```python
topic_ids = df_clustered['topic_id'].unique()

topic_ids
```

```python
cluster_mapping = dict()
```

```python
for topic_id in topic_ids:
    games_list = df_clustered[df_clustered['topic_id'] == topic_id]['game'].values
    cur_prompt = prompt_template.format(list_of_games = games_list)

    response = await async_model_client.chat.completions.create(
        messages=[{"role": "user", "content": cur_prompt}], 
        model=model
    )
    
    answer = response.choices[0].message.content

    print(answer)
    
    cluster_mapping[topic_id] = answer
    
```

```python
cluster_mapping
```

```python
df_clustered['cluster_name'] = df_clustered['topic_id'].map(cluster_mapping)
```

```python
df_clustered[['game', 'cluster_name']].to_csv('tag_based_clustering')
```
