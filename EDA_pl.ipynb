### Данные с парсера, запущенного на польский стим

```python
import pandas as pd
df = pd.read_csv('result_pl.csv')
df
```

```python
# Информация о данных
print("\nИнформация о данных:")
print(df.info())
```

```python

# Проверка на пропущенные значения
print("\nКоличество пропущенных значений в каждом столбце:")
print(df.isnull().sum())
```

```python

# Базовая статистика
print("\nБазовая статистика числовых столбцов:")
df.describe()
```

```python
# Уникальные значения в каждом столбце
print("\nКоличество уникальных значений в каждом столбце:")
print(df.nunique())
```

```python
# Dыбрось все строки, где в столбце 'price' есть пропущенные значения
df = df.dropna(subset=['price'])
# Убери все дублирующиеся строки
df = df.drop_duplicates()
```

```python
df
```

```python
# Дубликаты в столбце 'name'
duplicates = df[df.duplicated(subset=['name'], keep=False)]
duplicates
```

```python
# Fill 'price' with 0 if it contains "Free" or "Free to Play"
# df.loc[df['price'].str.contains('Free', na=False), 'price'] = '0 руб.'
df.loc[df['price'].str.contains('Free', na=False), 'price'] = '0zł'
df
```

```python

# Remove " руб." and replace commas with dots
# df['price'] = df['price'].str.replace(' руб.', '', regex=False).replace(',', '.', regex=False)
df['price'] = (df['price'].str.replace('zł', '', regex=False)).str.replace(',', '.', regex=False)

# Convert 'price' to numeric
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df = df.dropna(subset=['price'])
# Создай новую колонку, где будет добавлена Версия игры на основе сортировки по 'price'
df['version'] = df.groupby('name')['price'].rank(method='first', ascending=True).astype(int)
# df = df.sort_values('price').drop_duplicates(subset=['name'], keep='first')
```

```python
df
```

```python
df.sort_values('price', ascending=True, inplace=True)
df
```

```python
# Преобразуй reviews_numbers в числовой тип
df['reviews_numbers'] = pd.to_numeric(df['reviews_numbers'], errors='coerce')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Correlation matrix
correlation_matrix = numeric_df.corr()
print("\nCorrelation matrix:")
correlation_matrix
```

```python
# Heatmap of correlations
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
for col in numerical_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    print(f"\nValue counts for {col}:")
    print(df[col].value_counts())
```

```python
for col in numerical_columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.show()
```

```python
if 'price' in df.columns and 'reviews_numbers' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=df['price'], y=df['reviews_numbers'])
    plt.title('Price vs Reviews Numbers')
    plt.xlabel('Price')
    plt.ylabel('Reviews Numbers')
    plt.show()
```
